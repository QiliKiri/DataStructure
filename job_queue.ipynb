```python
import math
from collections import deque
from collections import namedtuple
```

```python
AssignedJob = namedtuple("AssignedJob", ["worker", "started_at"])
```

```python
class Job_queue():
    def __init__(self, workers, jobs):
        self.workers = workers
        self.jobs = jobs
        self.job_i = 0
        self.q = []
        self.ans = []
        self.flag = False
            
        if self.workers >= len(self.jobs):
            for i in range(len(self.jobs)):
                self.ans.append(AssignedJob(i, 0))
            self.flag = True
        else:
            for i in range(self.workers):
                self.q.append( (i, 0, jobs[i]) )
                self.ans.append(AssignedJob(i, 0))
                self.job_i += 1
        
    def initialize(self):
        i = len(self.q) - 1
        while i >= 0:
            self.sift_down(i)
            i -= 1
            
    def sift_down(self, i):
        size_i =len(self.q) - 1
        while True:
            l_i = 2*i + 1
            r_i = 2*i + 2
            
            # two children
            if r_i <= size_i:
                # parent in the middle, children on the sides
                # 3 3 3
                if self.q[i][2] == self.q[l_i][2] and self.q[l_i][2] == self.q[r_i][2]:
                    the_min = min(self.q[i][0], self.q[l_i][0], self.q[r_i][0])
                    if the_min == self.q[i][0]:
                        break
                    elif the_min == self.q[l_i][0]:
                        self.q[i], self.q[l_i] = self.q[l_i], self.q[i]
                        i = l_i
                    elif the_min == self.q[r_i][0]:
                        self.q[i], self.q[r_i] = self.q[r_i], self.q[i]
                        i = r_i
                
                # 1/2 3 1/2
                elif self.q[i][2] > self.q[l_i][2] and self.q[i][2] > self.q[r_i][2]:
                    # 1 3 1
                    if self.q[l_i][2] == self.q[r_i][2]:
                        if self.q[l_i][0] <= self.q[r_i][0]:
                            self.q[i], self.q[l_i] = self.q[l_i], self.q[i]
                            i = l_i
                        else:
                            self.q[i], self.q[r_i] = self.q[r_i], self.q[i]
                            i = r_i
                    # 1 3 2
                    elif self.q[l_i][2] < self.q[r_i][2]:
                        self.q[i], self.q[l_i] = self.q[l_i], self.q[i]
                        i = l_i
                    # 2 3 1
                    elif self.q[r_i][2] < self.q[l_i][2]:
                        self.q[i], self.q[r_i] = self.q[r_i], self.q[i]
                        i = r_i
                        
                # 2 1 3
                elif self.q[i][2] < self.q[l_i][2] and self.q[i][2] < self.q[r_i][2]:
                    break
                    
                # 1 2 2/3
                elif self.q[i][2] > self.q[l_i][2] and self.q[i][2] <= self.q[r_i][2]:
                    self.q[i], self.q[l_i] = self.q[l_i], self.q[i]
                    i = l_i
                
                # 2/3 2 1
                elif self.q[i][2] > self.q[r_i][2] and self.q[i][2] <= self.q[l_i][2]:
                    self.q[i], self.q[r_i] = self.q[r_i], self.q[i]
                    i = r_i   
                    
                # 1 1 2
                elif self.q[i][2] == self.q[l_i][2] and self.q[i][2] < self.q[r_i][2]:
                    if self.q[i][0] > self.q[l_i][0]:
                        self.q[i], self.q[l_i] = self.q[l_i], self.q[i]
                        i = l_i
                    else:
                        break
                
                # 2 1 1
                elif self.q[i][2] == self.q[r_i][2] and self.q[i][2] < self.q[l_i][2]:
                    if self.q[i][0] > self.q[r_i][0]:
                        self.q[i], self.q[r_i] = self.q[r_i], self.q[i]
                        i = r_i
                    else:
                        break
            
            # one child
            elif l_i <= size_i:
                # 1 2
                if self.q[i][2] > self.q[l_i][2]:
                    self.q[i], self.q[l_i] = self.q[l_i], self.q[i]
                    i = l_i
                # 1 1
                elif self.q[i][2] == self.q[l_i][2]:
                    if self.q[i][0] > self.q[l_i][0]:
                        self.q[i], self.q[l_i] = self.q[l_i], self.q[i]
                        i = l_i
                    else:
                        break
                # 2 1
                else:
                    break   
                    
            # no child      
            else:
                break
                
    def job_processing(self):
        if self.flag:
            pass
        else:
            self.initialize()
            while self.job_i < len(self.jobs):
                worker, started_at, finished_at = self.q[0][0], self.q[0][1], self.q[0][2]
                self.q.append( (worker, finished_at, self.jobs[self.job_i]+finished_at) )
                self.job_i += 1
                self.q[0], self.q[-1] = self.q[-1], self.q[0]
                self.q.pop()
                self.sift_down(0)
                self.ans.append(AssignedJob(worker, finished_at))
```

```python
jobs = [1,2,3,4,5]
workers = 2
a = Job_queue(workers, jobs)
a.job_processing()
a.ans
```
